# Exercise 28
#Define function percent(word,text) that calculates how often a given word occurs in a text and expresse the result as a percentage

```python
def percentage(word,text):
    contador = 0
    for w in text:
        if word in w:
            contador  = contador +1
    print(contador)

    lt = len(text)
    print(lt)
    percentage = (100.0*contador)/lt
    return percentage


text = ['Hi','this','is','a','proof','this','is','not','example']
per = percentage('this',text)
print('Percentage of the word this: ' + str(per) + '%')
```
